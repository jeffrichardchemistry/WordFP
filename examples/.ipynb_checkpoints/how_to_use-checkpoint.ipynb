```python
from WordFP import WordFP
```

```python
word = 'animal'
```

```python
#Instancy wordFP class
wfp = WordFP(max_word_length=23)
help(wfp) #internal documentation
```

```python
#This function returns de word encoding in matrix form
word_as_matrix = wfp.fit(word=word, engine='F', matrix_dtype='uint8')
print('animal\n',word_as_matrix)
```

```python
#transform wordFP in 1D array
word_as_matrix.ravel()
```

### Search similarity words

```python
#Fingerprints are in ravel form (array 1D)
database_just_FP = wfp.getbase_wordFP() #get database 217000 words PT-BR in ASCC
database_just_FP
```

```python
database_full = wfp.getbase_wordFP_full() #Fingerprints are in ravel form (array 1D)
database_full
```

```python
#This words are portuguese-Brazil
wfp.fit_similarity(base_wordFP=database_just_FP,
                   wordFP='animal',
                   complete_base=database_full, #dataframe with first column of words
                   similarity_metric='geometric', alpha=1, beta=1,
                   threshold=0.75)
```
